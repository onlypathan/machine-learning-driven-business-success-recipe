```python
import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Connect to SQLite database
conn = sqlite3.connect("../db/database_sqlite.db")

# Query numeric fields for correlation analysis
query = """
SELECT
    Rating,
    Score,
    Review_Count,
    Median_Household_Income,
    Median_Home_Value,
    Population_Density_per_sq_mi,
    Employment_Rate,
    Total_Employer_Establishments
FROM Combined_Restaurant_View
WHERE Rating IS NOT NULL
  AND Score IS NOT NULL
  AND Review_Count IS NOT NULL
  AND Median_Household_Income IS NOT NULL
  AND Median_Home_Value IS NOT NULL
  AND Population_Density_per_sq_mi IS NOT NULL
  AND Employment_Rate IS NOT NULL
  AND Total_Employer_Establishments IS NOT NULL
"""
df_corr = pd.read_sql_query(query, conn)
conn.close()

# Rename columns for readability
df_corr.rename(columns={
    "Rating": "Rating",
    "Score": "Inspection Score",
    "Review_Count": "Review Count",
    "Median_Household_Income": "Median Household Income",
    "Median_Home_Value": "Median Home Value",
    "Population_Density_per_sq_mi": "Population Density",
    "Employment_Rate": "Employment Rate",
    "Total_Employer_Establishments": "Employer Establishments"
}, inplace=True)

# Compute correlation matrix
correlation_matrix = df_corr.corr()

# Plot and save heatmap
plt.figure(figsize=(10, 9))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("\nCorrelation Matrix of Key Restaurant Performance Variables", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("6.2. zip_correlation_matrix.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import sqlite3
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load ZIP-level aggregated data
conn = sqlite3.connect("../db/database_sqlite.db")
query = """
SELECT
    Zip_Code,
    AVG(Rating) AS avg_rating,
    AVG(Score) AS avg_score,
    AVG(Review_Count) AS avg_reviews,
    AVG(Median_Household_Income) AS avg_income,
    AVG(Median_Home_Value) AS avg_home_value,
    AVG(Population_Density_per_sq_mi) AS avg_density,
    AVG(Employment_Rate) AS avg_employment,
    AVG(Total_Employer_Establishments) AS avg_establishments
FROM Combined_Restaurant_View
GROUP BY Zip_Code
HAVING COUNT(*) >= 5
"""
df_zip = pd.read_sql_query(query, conn)
conn.close()

# Normalize features for clustering
features = [
    'avg_rating', 'avg_score', 'avg_reviews',
    'avg_income', 'avg_home_value', 'avg_density',
    'avg_employment', 'avg_establishments'
]
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_zip[features])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_zip['Cluster'] = kmeans.fit_predict(df_scaled)

# Map cluster labels
cluster_labels = {
    0: 'Underserved',
    1: 'Emerging',
    2: 'High Potential'
}
df_zip['Cluster_Label'] = df_zip['Cluster'].map(cluster_labels)

# Plot cluster distribution
plt.figure(figsize=(8, 5))
sns.countplot(
    data=df_zip,
    x='Cluster_Label',
    hue='Cluster_Label',
    palette='Set2',
    legend=False
)
plt.title('Distribution of ZIP Codes by Cluster Type')
plt.xlabel('\nCluster Category')
plt.ylabel('Number of ZIP Codes')
plt.tight_layout()
plt.savefig("6.2. zip_clusters_labeled.png", dpi=300)
plt.show()

# Show ZIP codes per cluster
print(df_zip.groupby('Cluster_Label')['Zip_Code'].apply(list))
print()
```
