```python
import sqlite3
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load aggregated ZIP-level data
conn = sqlite3.connect("../db/database_sqlite.db")
query = """
SELECT
    Zip_Code,
    AVG(Rating) AS avg_rating,
    AVG(Review_Count) AS avg_reviews,
    AVG(Score) AS avg_score,
    AVG(Median_Home_Value) AS avg_home_value,
    AVG(Population_Density_per_sq_mi) AS avg_density,
    AVG(Employment_Rate) AS avg_employment,
    AVG(Total_Employer_Establishments) AS avg_establishments
FROM Combined_Restaurant_View
GROUP BY Zip_Code
HAVING COUNT(*) >= 5
"""
zip_df = pd.read_sql_query(query, conn)
conn.close()

# Normalize selected features
features_to_normalize = [
    'avg_rating',
    'avg_score',
    'avg_reviews',
    'avg_establishments',
    'avg_home_value'
]

scaler = MinMaxScaler()
zip_df[[f'norm_{col[4:]}' for col in features_to_normalize]] = (
    scaler.fit_transform(zip_df[features_to_normalize])
)

# Compute opportunity score
zip_df['Opportunity_Score'] = (
    0.30 * zip_df['norm_rating'] +
    0.25 * zip_df['norm_score'] +
    0.20 * zip_df['norm_reviews'] +
    0.15 * zip_df['norm_establishments'] +
    0.10 * zip_df['norm_home_value']
)

# Rank ZIP codes
zip_df['Rank'] = zip_df['Opportunity_Score'].rank(ascending=False)

# Show top 10 ZIP codes
top_10 = zip_df.sort_values(by='Opportunity_Score', ascending=False).head(10)
print("ZIP Opportunity Scores (Top Features):")
display(top_10)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Colors for bars
colors = [
    'steelblue', 'darkorange', 'seagreen', 'crimson', 'purple',
    'gold', 'teal', 'orchid', 'sienna', 'navy'
]

# Create bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(top_10['Zip_Code'], top_10['Opportunity_Score'], color=colors)

# Chart labels and formatting
plt.title('\nTop 10 LA City ZIP Codes by Restaurant Opportunity Score', fontsize=14)
plt.xlabel('ZIP Code\n')
plt.ylabel('Opportunity Score')
plt.xticks(rotation=45)
plt.ylim(0, 1)

# Add rank and percentage labels
for i, bar in enumerate(bars):
    score = top_10.iloc[i]['Opportunity_Score']
    rank = i + 1
    label = f"#{rank}\n{score * 100:.1f}%"
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        bar.get_height() + 0.01,
        label,
        ha='center',
        fontsize=10
    )

plt.tight_layout()
plt.savefig("6.1. top_zip_opportunity_score.png", dpi=300)
plt.show()
```
