```python
import sqlite3
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from database
conn = sqlite3.connect("../db/database_sqlite.db")
query = """
SELECT
    Restaurant_Info_ID,
    Restaurant_Name,
    Rating,
    Review_Count,
    Score,
    Price,
    Total_Employer_Establishments,
    Restaurant_Address,
    Zip_Code,
    Median_Household_Income,
    Median_Home_Value,
    Population_Density_per_sq_mi,
    Employment_Rate
FROM Combined_Restaurant_View
WHERE Rating IS NOT NULL
  AND Review_Count IS NOT NULL
  AND Score IS NOT NULL
  AND Price IS NOT NULL
  AND Total_Employer_Establishments IS NOT NULL
  AND Restaurant_Address IS NOT NULL
  AND Zip_Code IS NOT NULL
  AND Median_Household_Income IS NOT NULL
  AND Median_Home_Value IS NOT NULL
  AND Population_Density_per_sq_mi IS NOT NULL
  AND Employment_Rate IS NOT NULL
"""
df_rest = pd.read_sql_query(query, conn)
conn.close()

# Convert price symbols to numeric values
price_map = {'$': 1, '$$': 2, '$$$': 3, '$$$$': 4}
df_rest['Price_Encoded'] = df_rest['Price'].map(price_map)

# Normalize features
scaler = MinMaxScaler()
df_rest[
    [
        'norm_rating', 'norm_reviews', 'norm_score', 'norm_price', 'norm_employers',
        'norm_income', 'norm_home_value', 'norm_pop_density', 'norm_employment_rate'
    ]
] = scaler.fit_transform(
    df_rest[
        [
            'Rating', 'Review_Count', 'Score', 'Price_Encoded',
            'Total_Employer_Establishments', 'Median_Household_Income',
            'Median_Home_Value', 'Population_Density_per_sq_mi', 'Employment_Rate'
        ]
    ]
)

# Compute success index
df_rest['Success_Index'] = (
    0.30 * df_rest['norm_rating'] +
    0.25 * df_rest['norm_reviews'] +
    0.20 * df_rest['norm_score'] +
    0.15 * df_rest['norm_price'] +
    0.10 * df_rest['norm_employers']
) * 100

# Create binary success label
threshold = df_rest['Success_Index'].quantile(0.7)
df_rest['Success_Label'] = (df_rest['Success_Index'] >= threshold).astype(int)

# Prepare train-test data
X = df_rest[
    [
        'norm_rating', 'norm_reviews', 'norm_score', 'norm_price', 'norm_employers',
        'norm_income', 'norm_home_value', 'norm_pop_density', 'norm_employment_rate'
    ]
]
y = df_rest['Success_Label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train Random Forest classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]

# Model evaluation
print("\nRandom Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Random Forest Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))

# Feature importance plot
importances = pd.Series(
    rf_model.feature_importances_,
    index=[
        'Rating',
        'Review Count',
        'Inspection Score',
        'Price Level',
        'Employer Establishments',
        'Median Income',
        'Median Home Value',
        'Population Density',
        'Employment Rate'
    ]
)

importances.sort_values().plot(
    kind='barh',
    title='\nFeature Importance (Random Forest)'
)
plt.xlabel('Importance Score\n')
plt.tight_layout()
plt.savefig("6.4. feature_importance_random_forest.png", dpi=300)
plt.show()

# ROC curve
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

plt.figure(figsize=(8, 6))
plt.plot(
    fpr_rf, tpr_rf,
    color='green',
    label=f'Random Forest ROC Curve (AUC = {roc_auc_rf:.2f})'
)
plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
plt.xlabel('False Positive Rate\n')
plt.ylabel('True Positive Rate')
plt.title('\nROC Curve - Random Forest')
plt.legend()
plt.tight_layout()
plt.savefig("6.4. roc_curve_random_forest.png", dpi=300)
plt.show()

# Boxplots by success label
df_rest['Success_Label_Cat'] = df_rest['Success_Label'].map({0: 'Low', 1: 'High'})

plot_features = {
    'norm_rating': 'Rating',
    'norm_reviews': 'Review Count',
    'norm_score': 'Inspection Score',
    'norm_price': 'Price Level',
    'norm_employers': 'Employer Establishments',
    'norm_income': 'Median Income',
    'norm_home_value': 'Median Home Value',
    'norm_pop_density': 'Population Density',
    'norm_employment_rate': 'Employment Rate'
}

fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.flatten()

for i, (feature, label) in enumerate(plot_features.items()):
    sns.boxplot(
        data=df_rest,
        x='Success_Label_Cat',
        y=feature,
        ax=axes[i],
        palette='pastel'
    )
    axes[i].set_title(f'\n{label} by Success Label')
    axes[i].set_xlabel('Success Label\n')
    axes[i].set_ylabel('Normalized ' + label)

plt.tight_layout()
plt.savefig("6.4. boxplots_3x3_success_factors.png", dpi=300)
plt.show()
```
